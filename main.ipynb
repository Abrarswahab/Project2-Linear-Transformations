```python
import cv2
import numpy as np
from google.colab import files
import matplotlib.pyplot as plt
from transformation.trans import Transformations
```

```python
# Specify the path to your image file
filename = "utilitis/tuwaiqimg.jpg"  # Replace with your image path
# Read the image
img_data = cv2.imread(filename)
# Convert from BGR (OpenCV default) to RGB (Matplotlib default)
img_rgb = cv2.cvtColor(img_data, cv2.COLOR_BGR2RGB)

# Initialize the Transformations object with the RGB image
trans = Transformations(img_rgb)
```

```python

# Open the file picker for the user to upload an image
uploaded = files.upload()
# Extract the name of the uploaded file
filename = list(uploaded.keys())[0]
# Read the image from the Colab storage
img_data = cv2.imread(filename)
# Convert from BGR (OpenCV default) to RGB (Matplotlib default)
img_rgb = cv2.cvtColor(img_data, cv2.COLOR_BGR2RGB)

# Initialize the Transformations object with the RGB image
trans = Transformations(img_rgb)
```

```python
# Scaling
# Prompt user for input values (Scale Factors)
scale_x = float(input("Enter Scale X value: "))
scale_y = float(input("Enter Scale Y value: "))
# Call the scaling method from our class

result = trans.scaling(scale_x, scale_y)
# Display the original vs the scaled image
trans.display(result, f"Scaled Image ({scale_x}x, {scale_y}y)")
```

```python
# Rotation
rotated_image = trans.rotate(90)
trans.display(rotated_image, f"Rotated Image")
```

```python
# Shear
# Prompt user for the shear factor (k). 0.5 is a good starting value
k_val = float(input("Enter Shear factor (k): "))

# Apply the shear transformation matrix to the image
sheared_result = trans.shear(k_val)

# Display the result (Note how the width increases to accommodate the slant)
trans.display(sheared_result, f"Sheared Image (k={k_val})")
```

```python
# Horizontal Reflection
# No user input needed as reflection is a fixed flip across the Y-axis
print("Applying Horizontal Reflection (Mirror Effect)...")

# Execute the horizontal reflection method
h_reflected = trans.reflect_horizontal()

# Display the mirror image
trans.display(h_reflected, "Horizontal Reflection (Flip X)")
```

```python
# Vertical Reflection
# This flips the image upside down across the X-axis
print("Applying Vertical Reflection (Upside Down Effect)...")

# Execute the vertical reflection method
v_reflected = trans.reflect_vertical()

# Display the flipped image
trans.display(v_reflected, "Vertical Reflection (Flip Y)")
```
